## Creating a file with the aggregation protocol applied

```python
# Function for aggregating the annotations
def add_tube_aggregation(df, column_name):
    aggre = []

    a1 = df[column_name+"_a1"]
    a2 = df[column_name+"_a2"]
    
    for i in range(len(df)):
        if a1[i] > 0.5 and a2[i] > 0.5:
            aggre.append(1)
        elif a1[i] < 0.5 and a2[i] < 0.5:
            aggre.append(0)
        else:
            aggre.append(-1)

    return aggre        
```

```python
# Adding the aggregated annotations to the df
annotations_save["Chest_drain_tube_aggregation"] = add_tube_aggregation(annotations_save, "Chest_drain_tube") 
annotations_save["NSG_tube_aggregation"] = add_tube_aggregation(annotations_save, "NSG_tube")
annotations_save["Endotracheal_tube_aggregation"] = add_tube_aggregation(annotations_save, "Endotracheal_tube")
annotations_save["Tracheostomy_tube_aggregation"] = add_tube_aggregation(annotations_save, "Tracheostomy_tube")
```

```python
# Adding a column with image paths
annotated_image_paths = []
for idx, row in annotations_save.iterrows():
    path = '../../../purrlab_students/padchest-preprocessed/' + str(row['ImageDir']) + '/' + str(row['ImageID'])
    annotated_image_paths.append(path)

annotations_save['ImagePath'] = annotated_image_paths
```

```python
# Extracting a subset of df columns
annotations_save = annotations_save[['ImageID', 'ImageDir', 'ImagePath', 'Chest_drain_tube_aggregation', 'NSG_tube_aggregation', 'Endotracheal_tube_aggregation', 'Tracheostomy_tube_aggregation']]

# Changing some column names
annotations_save.rename(columns = {'Chest_drain_tube_aggregation':'Chest_drain', 'NSG_tube_aggregation': 'NSG_tube', 'Endotracheal_tube_aggregation': 'Endotracheal_tube', 'Tracheostomy_tube_aggregation': 'Tracheostomy_tube'}, inplace = True)
```
